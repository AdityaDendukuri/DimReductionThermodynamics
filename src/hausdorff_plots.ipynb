```python
%matplotlib inline

from dataloader import *
from plotter import *
from neural_net import *
from scripts import *
from itertools import combinations 
```

```python
#load dataset
coordinates = load_features("../data/features.txt", 3)
potentials = load_labels("../data/labels.txt", 1)
num_data = len(coordinates)
num_particles = len(coordinates[0])/3 
```

```python
#extract some information from the feature space 
side_length_space = [] #side lengths between each particles
perimeter = [] #side lengths between each particles
n_bonds = np.zeros(len(coordinates)) #number of bonds in the structure
cut_off_distance = 1.6 #sigma  
for i in range(len(coordinates)):
    side_length_space.append([])
    particles = np.reshape(coordinates[i], [num_particles, 3])
    perms = combinations(particles, 2) 
    for perm in perms:
        side_length_space[i].append(np.linalg.norm(perm[0] - perm[1]))   
    perimeter.append(np.sum(side_length_space[i]))
    for j in range(len(side_length_space[i])):
        if side_length_space[i][j] < cut_off_distance:
            n_bonds[i] += 1 
```

```python
eigen = []
with open("../data/LJ3P_eigenvalues_HD.txt", 'r') as f:
    lines = f.readlines()
    for line in lines:
        data = line.split()
        eigen.append(np.array([float(data[1]), float(data[2]), float(data[3])]))
eigen = np.asarray(eigen)
plot_2d(eigen[:, 0], eigen[:, 2], color=n_bonds.flatten())
```

```python
rog = np.zeros(num_data)
with open("../data/rg_018.txt") as f:
    lines = f.readlines()
    for i, line in enumerate(lines):
        rog[i] = float(line)
plot_2d(eigen[:, 0], rog, color=n_bonds.flatten())
```

```python
plot_2d(eigen[:, 0], perimeter, color=n_bonds.flatten())
```

```python
area = []
for i in range(len(coordinates)):
    points = np.reshape(coordinates[i], [3, 3])
    r1 = points[1] - points[0]
    r2 = points[2] - points[0]
    area.append(np.linalg.norm(np.cross(r1, r2))*0.5)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
a = ax.scatter(eigen[:, 0], area, c=np.array(n_bonds).flatten())
ax.set_xlabel('reduced var 1')
ax.set_ylabel('area')
fig.colorbar(a, label=' n_bonds')
```
